```python
#pipelineتسمح بجمع وظائف معالجة اللغة المختلفة في خطوات متسلسلة
#set_seedدالة من مكتبة transformers تُستخدم للتحكم في سلوك العشوائية في نماذج المُحوِّل.
from transformers import pipeline, set_seed
#مكتبة  تُستخدم لتشكيل النصوص وتقسيمها إلى أسطر مُحددة
#توفر وظائف لقص النصوص الطويلة،وتغيير تنسيق النص
import textwrap
# تُستخدم للتعامل مع المصفوفات متعددة الأبعاد والعمليات الحسابية عليها
import numpy as np
#تُستخدم لإنشاء الرسومات البيانية والمخططات
import matplotlib.pyplot as ptl
#أداة من مكتبة البايثون تُستخدم لطباعة البيانات المُنسقة بشكل جميل.
#تُستخدم لعرض محتويات المصفوفات والقواميس والقوائم بشكل مُقرأ
from pprint import pprint
```

```python
#لعرض محتوى ملف نصي
!cat robert_forst.txt
```

```python
#إزالة المسافات من نهاية كل السلسة في الملف النصي
lines = [line.rstrip( ) for line in open('robert_forst.txt')]
lines=[line for line in lines if len(line) > 0]
```

```python
#مودل يعمل على كتابة جمل لاكمال فقرة او فقرات لإكمال مقال او فصول لإكمال رواية
gen = pipeline("text-generation")
```

```python
#دالة لتولد الارقام العشوائية بقيم اولية قمتها 1234
set_seed(1234)
```

```python
lines[0]
```

```python
#توليد نص جديد في كل تنفيذ
out=gen(lines[0])
```

```python
pprint(_)
```

```python
#من اجل الحصول عل محتوى القائمه
pprint(out[0]['generated_text'])
```

```python
#طباعة نص لا يزيد عن 20 كلمه
pprint(gen(lines[0], max_length=20))
```

```python
#توليد بشكل متعددواختار واحد
out = gen(lines[0], num_return_sequences=3, max_length=20)
pprint(out)
```

```python
#دالة تعمل على ايجاد الكلمة المفقوده
def wrap(x):
  return textwrap.fill(x,replace_whitespace=False,fix_sentence_endings=True)
```

```python
#استدعاء الداله Wrap
out=gen(lines[0],max_length=30)
print(wrap(out[0]['generated_text']))
```

```python
prev = "Two roads diverged in a yellow wood, including one that blocked the" + \
    " road leading to another intersection in the middle of the city."

out = gen(prev+ '\n' + lines[2], max_length=60)
print(wrap(out[0]['generated_text']))
```

```python
prompt = "Newral networks with attention have been used with great success" + \
    " in natural language processing"

out = gen(prompt, max_length=300)
print(wrap(out[0]['generated_text']))
```
